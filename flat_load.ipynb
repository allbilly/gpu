```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./flat_load_gold.csv")
```

```python
df
```

```python
grouped_df = df.groupby('param1')
mean_df = grouped_df.mean()
```

```python
mean_df
```

```python
linear_regressor = LinearRegression()
```

```python
x = mean_df.index.values.reshape(-1, 1)
linear_regressor.fit(x,mean_df['time'].values)
time_predict = linear_regressor.predict(x)
print(linear_regressor.coef_)
```

```python
plt.plot(x,time_predict,mean_df['time'])
```
