```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./result.csv')
data['time']=data['time']*10**9
```

```python
data
```

```python
kb = pd.DataFrame(columns=['k','b'])
```

```python
df = data.pivot(index = 'param1', columns = 'param2', values = 'time')

linear_regressor = LinearRegression()
x = df.index.values.reshape(-1, 1)
fig = plt.figure(figsize=(20,10))
for col in df.columns:
    linear_regressor.fit(x,df[col].values)
    entry = [linear_regressor.coef_,linear_regressor.intercept_]
    kb.loc[len(kb)] = entry
    time_pred = linear_regressor.predict(x)
    plt.plot(x,time_pred,label='wf='+str(col))
    plt.xlim(0,16384)
    plt.legend()
```
